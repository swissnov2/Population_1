___Bar chart___

```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

df = pd.read_csv("F:\Population.csv")

country_data = df[df["Country Name"].isin(["Argentina", "Aruba", "Barbados"])]

plt.bar(
    country_data["Country Name"],
    country_data["2019 [YR2019]"]
)

plt.title("Population in 2019")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```

___Histogram___

```python
plt.hist(df["2019 [YR2019]"])

plt.title("Population Distribution")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

___Scatter chart___

```python
plt.scatter(
    df["2010 [YR2010]"],
    df["2019 [YR2019]"]
)

plt.xlabel("Population 2010")
plt.ylabel("Population 2019")
plt.show()
```

___Pie chart___

```python
sizes = [40, 30, 20, 10]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels)

plt.show()
```

___Box chart___

```python
plt.boxplot(df["2019 [YR2019]"])

plt.title("Population Outliers")
plt.show()
```
